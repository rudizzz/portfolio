# Dataset Description

This dataset consists of more than 600 streamers (Indonesia, Vietnam, Philippines). Each row represents data of one unique streamer. There are different sets of facial features extracted using an AI tool (e.g., Self_Esteem_, Character_Facet_, Personal_Values_Facet_, Temperament_), in addition to Gender, Game, Country, Total Follower, and the role they play in the game (e.g., Role_Director).

**Task**: find the features that have the most influence on the dependent variable (PaidStarPerWatchedHour)

# Data Cleaning and Manipulation

## Drop duplicate features

```python
df = df.drop(columns='PaidStarPerWatchedHour.1')
```

## Remove missing values

```python
df_null = df[df.isnull().any(axis=1)]
print(df_null.shape)
```

```python
df_miss = pd.concat([df_null, df[df.values == "-"]])
print(df_miss.shape)
```

```python
df = pd.concat([df, df_miss]).drop_duplicates(keep=False)
df
```

## Correct the incorrect value names

```python
game_name = list(df['Game'].unique())
game_name.sort()
game_name
```

```python
df['Game'] = df['Game'].replace(['Agge of Empires'], 'Age of Empires')
```

```python
total_follower = df['Total Follower'].apply(lambda x: x.replace(',',''))
total_follower = total_follower.astype(int)
df['Total Follower'] = total_follower
```

```python
df.describe()
```

## Dealing with Outliers

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.scatterplot(x='Broadcast Hours', y='PaidStarPerWatchedHour', data=df)
plt.title('Broadcast Hours x PaidStarPerWatchedHour', y=1.05)
plt.show()
```

```python
q1_broadcast = np.quantile(df['Broadcast Hours'], 0.25)
q3_broadcast = np.quantile(df['Broadcast Hours'], 0.75)
#iqr_broadcast = q3_broadcast - q1_broadcast
iqr_broadcast = iqr(df['Broadcast Hours'])
```

```python
lower_limit_broadcast = q1_broadcast - 1.5 * iqr_broadcast
upper_limit_broadcast = q3_broadcast + 1.5 * iqr_broadcast
broadcast_outliers = df[(df['Broadcast Hours'] < lower_limit_broadcast) | (df['Broadcast Hours'] > upper_limit_broadcast)]
print(broadcast_outliers.shape)
```

```python
df = pd.concat([df,broadcast_outliers]).drop_duplicates(keep=False)
print(df.shape)
```

```python
q1_paidstar = np.quantile(df['PaidStarPerWatchedHour'], 0.25)
q3_paidstar = np.quantile(df['PaidStarPerWatchedHour'], 0.75)
iqr_paidstar = iqr(df['PaidStarPerWatchedHour'])
```

```python
lower_limit_paidstar = q1_paidstar - 1.5 * iqr_paidstar
upper_limit_paidstar = q3_paidstar + 1.5 * iqr_paidstar
paidstar_outliers = df[(df['PaidStarPerWatchedHour'] < lower_limit_paidstar) | (df['PaidStarPerWatchedHour'] > upper_limit_paidstar)]
print(paidstar_outliers.shape)
```

```python
df = pd.concat([df,paidstar_outliers]).drop_duplicates(keep=False)
print(df.shape)
```

```python
q1_follower = np.quantile(df['Total Follower'], 0.25)
q3_follower = np.quantile(df['Total Follower'], 0.75)
iqr_follower = iqr(df['Total Follower'])
```

```python
lower_limit_follower = q1_follower - 1.5 * iqr_follower
upper_limit_follower = q3_follower + 1.5 * iqr_follower
follower_outliers = df[(df['Total Follower'] < lower_limit_follower) | (df['Total Follower'] > upper_limit_follower)]
print(follower_outliers.shape)
```

```python
new_df = pd.concat([df, follower_outliers]).drop_duplicates(keep=False)
print(new_df.shape)
```

# Build and Run Machine Learning Models
to find wich features will make the streamers more or less likely to receive stars from their audience, in this case PaidStarPerWatchedHour is our dependent variable.  

```python
X = new_df.drop(columns=['Country','Gender','Game','MBTI_grouped','PaidStarPerWatchedHour', 'Total Follower log', 'Broadcast Hours log'])
y = new_df['PaidStarPerWatchedHour']
```

Normalization (min-max scalar)

```python
X['Total Follower'] = (X['Total Follower']-X['Total Follower'].min()) / (X['Total Follower'].max()-X['Total Follower'].min())
X['Broadcast Hours'] = (X['Broadcast Hours']-X['Broadcast Hours'].min()) / (X['Broadcast Hours'].max()-X['Broadcast Hours'].min())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
sort = rf.feature_importances_.argsort()
fig, ax = plt.subplots(figsize=(15,30))
plt.barh(X.columns[sort], rf.feature_importances_[sort])
plt.xlabel('Random Forest Feature Importance')
plt.show()
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
```

```python
dt.feature_importances_
```

```python
sort = dt.feature_importances_.argsort()
fig, ax = plt.subplots(figsize=(15,30))
plt.barh(X.columns[sort], dt.feature_importances_[sort])
plt.xlabel('Decision Tree Feature Importance')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
sort = lr.coef_.argsort()
fig, ax = plt.subplots(figsize=(15,30))
plt.barh(X.columns[sort], lr.coef_[sort])
plt.xlabel('Linaer Regression Feature Importance')
```

# The Result

Based on the results, Broadcast Hours and Total Follower are the features that make streamers more likely to receive stars from viewers (in the Random Forest and Decision Tree Model). Unlike other algorithms, Linear Regression shows us that the features extracted with ai, namely Temprament_Sanguine and Temprament_Centric are the most features that make streamers more likely to receive stars.
